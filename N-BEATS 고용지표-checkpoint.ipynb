```python
import pandas as pd
import matplotlib.pyplot as plt
from darts import TimeSeries
from darts.models import NBEATSModel
from darts.dataprocessing.transformers import Scaler
from sklearn.preprocessing import MinMaxScaler
from darts.metrics import mape
```

```python
# 1.  
df = pd.read_csv(".csv", header=1)
df[''] = pd.to_datetime(df[''].str.strip().str.replace(" ", "").str.replace(".", "-"))
df.set_index('', inplace=True)

# 2. () 
ts_data = df['.1']

# 3. TimeSeries  
series = TimeSeries.from_series(ts_data)

# 4. 
scaler = Scaler(MinMaxScaler())
series_scaled = scaler.fit_transform(series)

# 5. /  
train, val = series_scaled[:-12], series_scaled[-12:]

# 6. N-BEATS  
model = NBEATSModel(
    input_chunk_length=24,
    output_chunk_length=12,
    n_epochs=300,
    random_state=42,
   generic_architecture=True,  #    
    num_blocks=1,  #    : 1
    num_layers=4,  #    : 4
    layer_widths=512,  #   : 512 
    trend_polynomial_degree=2,  #   : 2
    batch_size=800,  #  : 800
    dropout=0.0,
)

# 7.  
model.fit(train, verbose=True)

# 8.  
forecast = model.predict(n=12)

# 9.   
forecast_unscaled = scaler.inverse_transform(forecast)
train_unscaled = scaler.inverse_transform(train)
val_unscaled = scaler.inverse_transform(val)

# 10. MAPE 
error = mape(val_unscaled, forecast_unscaled)
print(f"MAPE: {error:.2f}%")

# 11. 
plt.figure(figsize=(10, 5))
series.plot(label=" ")
train_unscaled.plot(label=" ")
val_unscaled.plot(label="")
forecast_unscaled.plot(label="")
plt.title("N-BEATS () ")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()




```
