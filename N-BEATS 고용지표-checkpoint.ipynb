```python
import pandas as pd
import matplotlib.pyplot as plt
from darts import TimeSeries
from darts.models import NBEATSModel
from darts.dataprocessing.transformers import Scaler
from sklearn.preprocessing import MinMaxScaler
from darts.metrics import mape
```

```python
# 1. 데이터 불러오기
df = pd.read_csv("고용지표.csv", header=1)
df['시점'] = pd.to_datetime(df['시점'].str.strip().str.replace(" ", "").str.replace(".", "-"))
df.set_index('시점', inplace=True)

# 2. 실업율(합계)만 선택
ts_data = df['합계.1']

# 3. TimeSeries 객체로 변환
series = TimeSeries.from_series(ts_data)

# 4. 스케일링
scaler = Scaler(MinMaxScaler())
series_scaled = scaler.fit_transform(series)

# 5. 학습/테스트 데이터 분할
train, val = series_scaled[:-12], series_scaled[-12:]

# 6. N-BEATS 모델 정의
model = NBEATSModel(
    input_chunk_length=24,
    output_chunk_length=12,
    n_epochs=300,
    random_state=42,
   generic_architecture=True,  # 일반 아키텍처 사용 
    num_blocks=1,  # 각 스택당 블록 수: 1개
    num_layers=4,  # 각 블록당 레이어 수: 4개
    layer_widths=512,  # 각 레이어의 너비: 512 노드
    trend_polynomial_degree=2,  # 추세 다항식 차수: 2차
    batch_size=800,  # 배치 크기: 800
    dropout=0.0,
)

# 7. 모델 학습
model.fit(train, verbose=True)

# 8. 예측 수행
forecast = model.predict(n=12)

# 9. 예측 결과 역변환
forecast_unscaled = scaler.inverse_transform(forecast)
train_unscaled = scaler.inverse_transform(train)
val_unscaled = scaler.inverse_transform(val)

# 10. MAPE 출력
error = mape(val_unscaled, forecast_unscaled)
print(f"MAPE: {error:.2f}%")

# 11. 시각화
plt.figure(figsize=(10, 5))
series.plot(label="전체 시계열")
train_unscaled.plot(label="학습 데이터")
val_unscaled.plot(label="실제값")
forecast_unscaled.plot(label="예측값")
plt.title("N-BEATS 실업률(합계) 예측")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()




```
